5. What is mutability in Python ?

    In Python, mutability refers to the ability of an object to be changed  
    after it is created. An object is considered mutable if its internal state
    can be modified in place without creating a completely new object.
    Conversely, an object is immutable if its internal state cannot be altered
    after creation. Any operation that appears to modify an immutable object
    actually creates a new object with the changes.

6. Why are lists mutable, but tuples are immutable ?

    The choice between making lists mutable and tuples immutable was a
    deliberate design decision in Python, driven by the different purposes  
    these data structures are intended to serve. Lists provide flexibility for
    dynamic collections, while tuples offer data integrity, can be used as
    dictionary keys/set elements, and are suitable for representing fixed
    collections.

 This distinction allows Python to be a versatile language with data structures optimized for various programming tasks.

9 . What is type casting in python ?

    Type casting, also known as type conversion, is the process of changing an
    object from one data type to another. Python is a dynamically-typed
    language, which means you don't explicitly declare the data type of a
    variable when you create it. However, there are times when you need to
    convert a value from one type to another to perform certain operations or
    meet specific requirements.

    Python provides built-in functions to perform type casting. These functions attempt to create a new object of the target type based on the value of the original object.

5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.



```python
# Creating a list
my_list = [1, 2, 3, 4, 5]
print(f"Original list: {my_list}")

# Changing an element in the list
my_list[2] = 10
print(f"Modified list: {my_list}")

# Creating a tuple
my_tuple = (1, 2, 3, 4, 5)
print(f"\nOriginal tuple: {my_tuple}")

# Attempting to change an element in the tuple
try:
    my_tuple[2] = 10
except TypeError as e:
    print(f"Error when trying to modify tuple: {e}")
```

6. Write a function to demonstrate the behavior of mutable and immutable arguments.



```python
def modify_list(my_list):
    print(f"Original list inside function: {my_list}")
    my_list.append(100)
    print(f"Modified list inside function: {my_list}")

def modify_integer(my_int):
    print(f"Original integer inside function: {my_int}")
    my_int += 100
    print(f"Modified integer inside function: {my_int}")

# Demonstrate with a mutable argument (list)
my_list = [1, 2, 3]
print(f"Original list before function call: {my_list}")
modify_list(my_list)
print(f"List after function call: {my_list}")

# Demonstrate with an immutable argument (integer)
my_int = 10
print(f"\nOriginal integer before function call: {my_int}")
modify_integer(my_int)
print(f"Integer after function call: {my_int}")
```

9. Write a Python program to convert user input from string to integer, float, and boolean types.

```python
def convert_input(user_input):
    # Convert to integer
    try:
        int_value = int(user_input)
        print(f"Converted to integer: {int_value}")
    except ValueError:
        print("Cannot convert to integer")

    # Convert to float
    try:
        float_value = float(user_input)
        print(f"Converted to float: {float_value}")
    except ValueError:
        print("Cannot convert to float")

    # Convert to boolean
    if user_input.lower() in ['true', '1', 'yes', 'y']:
        bool_value = True
    elif user_input.lower() in ['false', '0', 'no', 'n']:
        bool_value = False
    else:
        bool_value = None
        print("Cannot convert to boolean")

    if bool_value is not None:
        print(f"Converted to boolean: {bool_value}")

# Get user input
user_input = input("Enter a value: ")
convert_input(user_input)
```

10. Write code to demonstrate type casting with list elements.




```python
my_list = ["1", "2", "3.14", "True", "False"]

# Type cast elements to integers
int_list = [int(x) for x in my_list[0:2]]
print("Integer list:", int_list)

# Type cast elements to float
float_list = [float(x) for x in my_list[2:3]]
print("Float list:", float_list)

# Type cast elements to boolean
bool_list = [bool(x) for x in my_list[3:5]]
print("Boolean list:", bool_list)

# Original list remains unchanged
print("Original list:", my_list)
```

15.  Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
def factorial(n):
    result = 1
    while n > 1:
        result *= n
        n -= 1
    return result

# Get user input
user_input = int(input("Enter a number to calculate its factorial: "))

# Calculate factorial
factorial_result = factorial(user_input)

# Print the result
print(f"The factorial of {user_input} is {factorial_result}")
```
